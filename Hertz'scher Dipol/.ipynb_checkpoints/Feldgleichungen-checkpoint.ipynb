```python
# Symbole
x, y, z, p1, p2, p3, t = symbols("x y z p1 p2 p3 t")

# Gleichungen
E1 = 3*(x*p1 + y*p2 + z*p3)

Ex = exp(-t) * (p1*(z**2 + y**2) - p3*x*z - p2*x*y + 3*E1*x - p1)
Ey = exp(-t) * (p2*(x**2 + z**2) - p1*x*y - p3*y*z + 3*E1*y - p1)
Ez = exp(-t) * (p3*(y**2 + x**2) - p2*z*y - p1*z*x + 3*E1*z - p1)

# Betrag des E-Feldes
Etotal = sqrt(Ex**2 + Ey**2 + Ez**2)
Etotal
```

```python
from sympy import *
init_printing()

# Symbole
x, y, z, r, p1, p2, p3, t, omega, p, c, epsilon, pi = symbols("x y z r p1 p2 p3 t omega p c epsilon pi ")

# Gleichungen
Skalierung = omega**3 / (4 * pi * epsilon * c**3)
p = omega * r /c
Welle = cos(omega*(r/c -t))
E1 = 3*(x*p1 + y*p2 + z*p3)

Ex2 = ((p1*(z**2 + y**2)  - p3*x*z - p2*x*y)**2 / p**2 + (3*E1*x - p1)**2*(1/p**6 -1/p**4))
Ey2 = ((p2*(x**2 + z**2) - p1*x*y - p3*y*z)**2 / p**2 + (3*E1*y - p1)**2*(1/p**6 -1/p**4))
Ez2 = ((p3*(y**2 + x**2) - p2*z*y - p1*z*x)**2 / p**2  + (3*E1*z - p1)**2*(1/p**6 -1/p**4))

Etotal = Skalierung * Welle * sqrt(Ex2 * Ey2 + Ez2)
print(Etotal)
Etotal
```
