## Neutronenstrahl in der Reaktorwand

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reaktorwandbereich / Schilddicke
shield = 3

# Anzahl der Teilchen
particles = 1000000
# Zeitlaufdauer
maxZeit  = 20
# Bewegungsgeschwindigkeit
speed = 1


pxpos = []
pypos = []
for p in range(particles):
    # Start und vergangene Punkte des Partikels
    x, y = 0, np.random.randn()
    pastx = [x]
    pasty = [y]
    
    # Lauf
    for frame in range(maxZeit):
        # Teilchen ist im Reaktor
        if pastx[frame] < 0:
            pastx.append(pastx[frame])
            pasty.append(pasty[frame])          
        # Teilchen ist aus dem Reaktor entkommen
        elif pastx[frame] >= shield:
            pastx.append(pastx[frame])
            pasty.append(pasty[frame])
        # Teilchen ist in der Wand
        else:
            pastx.append(pastx[frame] + (speed*np.random.random())*2-1), pasty.append(pasty[frame] + (speed*np.random.random())*2-1)
    
    # Endpunkt
    pxpos.append(pastx[maxZeit]); pypos.append(pasty[maxZeit])

        
plt.style.use("default")
plt.style.use("seaborn-dark")
plt.style.use("grayscale")
fig = plt.figure(figsize=(7,7), dpi=300)
fig.patch.set_facecolor('white')
plt.yticks([])

plt.plot(pxpos, pypos, "o", color=(0,0,0,0.2), markersize=1)
```
