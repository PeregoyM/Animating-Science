```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use("default")
plt.style.use("seaborn-dark")
plt.style.use("grayscale")

fig = plt.figure(figsize=(7,7), dpi=300)
fig.patch.set_facecolor('white')
plt.xticks([])
plt.yticks([])

# Startparameter
maxZeit  = 10000
# Bewegungsgeschwindigkeit
speed = 1

# Zeige die vergangenen Punkte
pastx = [0]
pasty = [0]

for frame in range(maxZeit):
    pastx.append(pastx[frame] + (speed*np.random.random())*2-1), pasty.append(pasty[frame] + (speed*np.random.random())*2-1)

plt.plot(pastx,pasty, lw=0.5)
```
