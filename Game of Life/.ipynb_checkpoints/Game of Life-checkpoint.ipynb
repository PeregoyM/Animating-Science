# Conway's Game of Life

```python
import numpy
import matplotlib.pyplot as plt
```

```python
# size of the env
size = 42
# time range
time = 20 * 1 #(fps * seconds)

# update rule to evolve the system
def play_life(a):
    xmax, ymax = a.shape
    b = a.copy()
    for x in range(xmax):
        for y in range(ymax):
            n = numpy.sum(a[max(x - 1, 0):min(x + 2, xmax), max(y - 1, 0):min(y + 2, ymax)]) - a[x, y]
            if a[x, y]:
                if n < 2 or n > 3:
                    b[x, y] = 0 # Rule 1 and 3
            elif n == 3:
                b[x, y] = 1 # Rule 4
    return(b)

# starting conditions
life = numpy.random.randint(2,size=(size,size))

# play conways game of life
for i in range(time):
    # show
    plt.style.use("default")
    plt.style.use("seaborn-dark")
    plt.style.use("grayscale")
    fig = plt.figure(figsize=(7,7), dpi=300)
    fig.patch.set_facecolor('white')
    plt.xticks([])
    plt.yticks([])
    plt.imshow(life, cmap="gray")
    plt.savefig("anim/conv_" + str(i) + ".png")
    # update
    life = play_life(life)
```
