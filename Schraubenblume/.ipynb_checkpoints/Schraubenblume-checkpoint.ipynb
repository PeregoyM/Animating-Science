```python
import numpy as np
import pylab as p
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as p3
%matplotlib notebook

res = 0.003
t = np.arange(0,3+res,res)
r = t * (1 + np.sin(16 * np.pi * t))
x = r * np.cos(2 * np.pi * t)
y = r * np.sin(2 * np.pi * t)
z = t

fig = p.figure(figsize=(8,8),dpi=120)
ax = p3.Axes3D(fig)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.plot(x,y,z, color="#618abf", lw=2)
p.show()
```
