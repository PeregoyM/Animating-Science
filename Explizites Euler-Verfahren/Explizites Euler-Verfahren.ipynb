# Explizites Euler-Verfahren

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import matplotlib.animation as animation
%matplotlib notebook
```

Es sei
$y' = 2y + x$ eine DGL erster Ordnung mit dem Anfangswert
$y(0) = 1$.

```python
# Schrittweite
h = 0.25 
# Startposition
x_0 = 0 
y_0 = 1

# 1. Schritt
y_1 = y_0 + (2*y_0 + x_0) * h
x_1 = x_0 + h 
print((x_1, y_1))
```

### Lösung

```python
def L(x):
    return 5/4*math.exp(2*x)-1/2*x-1/4

xt = np.arange(0,1,0.003)
yt = []
for e in xt:
    yt.append(L(e))
```

### Näherung

```python
y = [1]
x = [0]

plt.style.use("default")
plt.style.use("seaborn-dark")
plt.style.use("grayscale")
fig, ax = plt.subplots(figsize=(8,8),dpi=120) #figsize=(7,7)
fig.patch.set_facecolor('white')

Truth = plt.plot(xt,yt,"--", color="#048bf9", lw=2)
Euler, = ax.plot(x,y, "-o", color="#f9a004")

# Animiere die Planeten
def animate(i):
    y = [1]
    x = [0]
    h = 1/(i+1)
    it = 1*(i+1)
    for i in range(it):
        y.append(y[i] + (2 * y[i] + x[i]) * h)
        x.append(x[i]+h)
    
    Euler.set_data(x,y)

anim = animation.FuncAnimation(fig, animate, frames=100, interval=300)
print("Animation done.")
plt.show()
```
